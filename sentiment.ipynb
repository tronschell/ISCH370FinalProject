```python
import pandas as pd
import nltk
nltk.download("vader_lexicon")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
df = pd.read_csv("ISCHDATA1.csv", parse_dates=["date"])
df.dtypes
#df.head()
```

```python
sentiment_list = []

for i in df["tweet"]:
    sentiment_list.append(sia.polarity_scores(i)["compound"])

df["vader_sen"] = sentiment_list
df.head()
    
```

```python
test_df = df[["date", "vader_sen"]]

pd.to_datetime(test_df["date"])

test_df.set_index("date")

test_df.dtypes

computed_df = test_df.resample("D").sum().plot(kind="bar")
```
